```python
import plotly.graph_objs as go
from sklearn.datasets import make_regression
from IPython.display import Markdown as md
from plotly.offline import init_notebook_mode, iplot
import numpy as np

```

```python
init_notebook_mode(connected=True)
```

```python
def simple_linear_regression(x,y,z):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    mean_z = np.mean(z)
    theta2 = np.sum((y - mean_y)*(z - mean_z)) / np.sum(np.square(y-mean_y))
    theta1 = np.sum((x - mean_x)*(z - mean_z)) / np.sum(np.square(z-mean_z))
    theta0 = mean_z - theta1*mean_x -theta2*mean_y
    return theta0, theta1, theta2
```

```python
data = make_regression(200, n_features=2)
x = np.transpose(data[0])[0]
y = np.transpose(data[0])[1]
z = data[1]
```

```python
theta0, theta1, theta2 = simple_linear_regression(x,y,z)
t1 = range(-8,8)
t2 = range(-8,8)
xv, yv = np.meshgrid(t1, t2)
xvf = xv.flatten()
yvf = yv.flatten()
pred=theta0+theta1*xvf+theta2*yvf 
```

```python
reg_trace=go.Mesh3d(x=xvf,y=yvf,z=pred,color='red',opacity=0.5)
```

```python
data_trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    name='Input data'
)
```

```python
rmse = np.sqrt(np.mean((theta0+theta1*x+theta2*y-z)**2))
```

```python
md("Root-Mean Square Error = %r"%(rmse))
```

```python
fig = go.Figure(data=[data_trace, reg_trace])
iplot(fig)
```
